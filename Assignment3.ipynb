```python
class LL:
    def __init__(self):
        self.head = None
        self.tail = None

    class Node:
        def __init__(self, value):
            self.value = value
            self.next = None

    def insertNode(self, value):
        new_node = self.Node(value)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
            return
        self.tail.next = new_node
        self.tail = new_node

    def printLL(self):
        cnode = self.head
        while cnode:
            print(cnode.value, end=" ")
            cnode = cnode.next
        print()

    def rotateLL(self, k):
        if not self.head or k <= 0:
            return

        length = 1
        temp = self.head
        while temp.next:
            temp = temp.next
            length += 1

        k = k % length
        if k == 0:
            return

        temp.next = self.head
        count = length - k
        new_tail = self.head
        for _ in range(count - 1):
            new_tail = new_tail.next

        self.head = new_tail.next
        new_tail.next = None
        self.tail = new_tail

ll = LL()
values = [1, 2, 3, 4, 5]
for i in values:
    ll.insertNode(i)

ll.printLL()
ll.rotateLL(2)
ll.printLL()
```
